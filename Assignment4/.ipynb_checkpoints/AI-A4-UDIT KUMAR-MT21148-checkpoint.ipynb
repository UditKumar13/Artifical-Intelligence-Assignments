# Let us first preprocess the last column(prediction column)from our original data

```python
Y_Datta=job_Roledata.iloc[:,38]
```

```python
print("unique values in the last column is :",Y_Datta.nunique())
```

```python
#by examing I have decided to reduce the last column in 7 classes 
Y=Y_Datta.copy()
```

```python
def make_Reduction (uk_datta):
    for joob in range(uk_datta.shape[0]):
        if "Administrator" in uk_datta[joob]:
            uk_datta[joob] = 0
        elif "Analyst" in uk_datta[joob]:
            uk_datta[joob] = 1
        elif "Developer" in uk_datta[joob]:
            uk_datta[joob] = 2
        elif "Engineer" in uk_datta[joob]:
            uk_datta[joob] = 3
        elif "Architect" in uk_datta[joob]:
            uk_datta[joob] = 4
        elif "Manager" in uk_datta[joob]:
            uk_datta[joob] = 5
        else : 
            uk_datta[joob] = 6 
        
```

```python
#calling the reduction function on the last column 
#QA testing and support I have given the 7 number for it
make_Reduction(Y)
```

# Lets us try our first implementation of Ann to our data and Y prediction

```python
fina_Data_mlp = final_Data.copy()
Y_mlp = Y.copy()
final_Data = final_Data.values
```

```python
print("type before conversion : ",type(final_Data))
print("type after conversion : ",type(final_Data))
print(type(Y))
#Y=Y.values
print(type(Y))
```

```python

final_Data = np.asarray(final_Data).astype(np.float32)
Y = np.asarray(Y).astype(np.float32)

print(type(Y))
```

```python
#let us make the training and testing data 
#Train test split 
import random 
random.seed(3)
from sklearn.model_selection import train_test_split
#maintaining the data for the normal ascii values
datta_Xtrain, datta_Xtest, datta_Ytrain, datta_Ytest = train_test_split(final_Data, Y,test_size=0.3, random_state=42)
```

```python
print("shape of Xtrain : ",datta_Xtrain.shape)
print("shape of Xtest : ",datta_Xtest.shape)
print("shape of Ytrain : ",datta_Ytrain.shape)
print("shape of Ytest : ",datta_Ytest.shape)
```

# Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
datta_Xtrain = sc.fit_transform(datta_Xtrain)
datta_Xtest = sc.transform(datta_Xtest)
```

# Implementing the Ann using Keras

```python
from keras.models import Sequential
from keras import models
from keras import layers
from keras.layers.embeddings import Embedding
from keras.layers import Dense,Embedding,GRU,LSTM
import warnings
warnings.filterwarnings(action="ignore")
```

```python
uk_Classifier = models.Sequential()
#Input layer 
uk_Classifier.add(Dense(units = 38, activation = 'relu',input_shape=(38, )))
#hideen layer
uk_Classifier.add(Dense(units = 30, activation = 'relu'))
uk_Classifier.add(Dense(units = 1, activation = 'softmax'))
```

```python
uk_Classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics=['accuracy'])
```

```python
print("The summary of the classifer")
uk_Classifier.summary()
```

```python
uk_Fit = uk_Classifier.fit(datta_Xtrain,datta_Ytrain,batch_size=500,epochs=30,verbose=1,shuffle =True)
```

```python
y_pred = uk_Classifier.predict(datta_Xtest)
```

# Till now we got an accuracy of 18.5 percent on the Y test with 30-70 split

```python
from sklearn.metrics import accuracy_score
print("Accuracy on test set ", accuracy_score(y_pred,datta_Ytest))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(confusion_matrix(datta_Ytest, y_pred), annot=True, cmap='ocean_r')
plt.xlabel('Predicted')
plt.ylabel("Real")
plt.show()
```

```python
#Classifiaiton report
from sklearn.metrics import classification_report
print(classification_report(datta_Ytest, y_pred))
```
